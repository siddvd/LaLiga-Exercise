```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
data = pd.read_csv('Laliga.csv')
```

```python
data.info()
```

```python
#Renaming columns index with Row 0 for better understanding 
data.columns = data.loc[0] 
data2 = data.drop(0) #saving into new dataframe with dropped 1st row
data2.head(3) #preview data2 
```

```python
#replacing '-' with 0 in dataframe
data2.replace('-',0,inplace=True) 
```

```python
#to find out teams with Debut year 1930 to 1980
data2['Debut'] = data2['Debut'].astype(str).str[:4].astype(int)
Debut_year = data2[data2['Debut'].between(1930, 1980)]
Debut_year[['Team','Debut']]
```

```python
#Top 5 teams in terms of Points
data2['Points']= data2['Points'].astype(int)
data2[['Team','Points']].sort_values(['Points'], ascending=False).head(5)
```

```python
#Create function to know Goal difference

def Goal_diff_count(): 
    data2['Goal_dif'] = data2['GoalsFor'].astype(int) - data2['GoalsAgainst'].astype(int)
    return data2[['Team','Goal_dif']].sort_values(['Goal_dif'],ascending=False)
 
Goal_diff_count()
```

```python
#Maximum Goal difference
Goal_diff_count().head(1)
```

```python
#Minimum Goal difference
Goal_diff_count().tail(1)
```

```python
#New column winning percent 
data2['Winning_Percent'] = (data2['GamesWon'].astype(int)/data2['GamesPlayed'].astype(int))*100
```

```python
#Fill any numerical error column with 0 
data2['Winning_Percent'].fillna(0,inplace=True)
```

```python
#top 5 teams with winning percent 
data2[['Team','Winning_Percent']].sort_values('Winning_Percent',ascending=False).head(5)
```

```python
#Group teams based on Best Position
data2['BestPosition'] = data2['BestPosition'].astype(int)
data2[['Points','BestPosition']].groupby('BestPosition').sum()
```
